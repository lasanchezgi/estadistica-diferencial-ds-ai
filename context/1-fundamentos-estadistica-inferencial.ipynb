### Muestreo sistematico

```python
# Creando una función para el muestreo sistematico
def muestreo_sistematico(df,step):
    indexes = np.arange(0,len(df), step=step)
    muestra_sistematica = df.iloc[indexes]
    return muestra_sistematica
```

```python
muestreo_sistematico(df=df,step=50)
```

### Muestreo estratificado

Nuestros datos dicen que la proporción es la siguiente:
1. Hoteles en Cuauhtémoc: 0.5
2. Museos en Cuauhtémoc: 0.2
3. Hoteles en Venusiano Carranza: 0.1
4. Mercados en Cuauhtémoc: 0.1
5. Mercados en Venusiano Carranza: 0.1

```python
# Funciòn para el muestreo estratificado
def datos_estratificados(df, nombres_columnas_estratificadas, valores_estratificados, proporcion_estratificada, random_state=None):
    df_estratificado = pd.DataFrame(columns=df.columns)
    for i in range(len(valores_estratificados)):
        df_filtrado = df[df[nombres_columnas_estratificadas] == valores_estratificados[i]]
        ratio_len = int(len(df) * proporcion_estratificada[i])
        df_temporal = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state)
        df_estratificado = pd.concat([df_estratificado, df_temporal])

    return df_estratificado
```

```python
# Llama a la función pasando los parámetros adecuados
valores_estratificados = ['Cuautémoc, Hotel', 'Cuautémoc, Museo', 'Venustiano Carranza, Hotel', 'Cuauhtémoc, Mercado', 'Venustiano Carranza, Mercado']
proporcion_estratificada = [0.5, 0.2, 0.1, 0.1, 0.1]
```

```python
# Asegúrate de que el DataFrame 'df' esté definido antes de llamar a la función
# Llama a la función con los parámetros correctos
datos_estratificados(
    df=df,
    nombres_columnas_estratificadas='estratificado',  # Reemplaza 'estratificado' con el nombre de las columnas adecuadas
    valores_estratificados=valores_estratificados,
    proporcion_estratificada=proporcion_estratificada,
    random_state=42
)
```
